### Files
- to read file, scala File IO rely on `scala.io.Source`. it can read from file, string, url and stdin
- for reading binary file, write file or traversing directory, you need to use java class

### Process
- scala easily interact with shell script
- `scala.sys.process` package contains an implicit conversion from strings to `ProcessBuilder` objects. The `!` method executes the `ProcessBuilder` object.
- result of `!` is an exit code
- result of `!!` is a string


```scala211
import scala.io.Source
```

```scala211
val source = Source.fromFile("/home/alex/dev/servers/spark/LICENSE", "UTF-8")
for( l <- source.getLines()) println(l)
```
